```python
def solution(participant, completion):
    k = participant.copy()
    completion.sort()
    for a in range(len(participant)):
        for i in range(len(completion)):
            if participant[a]==completion[i]:
                k.remove(completion[i])
            else:
                pass
    answer = k
    return answer
```
